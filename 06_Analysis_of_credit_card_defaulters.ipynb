```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("./datasets/default_credit.xls")
df.head()
```

```python
#data preprocessing
print('SEX' + str(sorted(df['SEX'].unique())))
```

```python
fill = (df.EDUCATION == 0) | (df.EDUCATION == 5) | (df.EDUCATION == 6)
df.loc[fill, "EDUCATION"] = 4
```

```python
print('EDUCATION' + str(sorted(df['EDUCATION'].unique())))
```

```python
fill = (df.MARRIAGE == 0)
df.loc[fill, 'MARRIAGE'] = 2
```

```python
print('MARRIAGE' + str(sorted(df['MARRIAGE'].unique())))
```

```python
df = df.rename(columns={'default payment next month' : 'DEFAULT', 'PAY_0' : 'PAY_1'})
df.head()
```

```python
#Bivariate analysis
sns.set(rc={'figure.figsize':(15,10)})
edu = sns.countplot(x='SEX', hue='DEFAULT', data=df)
edu.set_xticklabels(['Male', 'Female'])
plt.show()
```

```python
pd.crosstab(df.SEX, df.DEFAULT, normalize='index', margins=True)
```

```python
#Evaluating the relationship between the default columns and the education and marriage columns
sns.set(rc={'figure.figsize': (15,10)})
edu = sns.countplot(x='EDUCATION', hue='DEFAULT', data=df)
edu.set_xticklabels(['Graduate School', 'University', 'High School', 'Other'])
plt.show()
```

```python
pd.crosstab(df.EDUCATION, df.DEFAULT, normalize=True)
```

```python
sns.set(rc={'figure.figsize':(12,10)})
marriage = sns.countplot(data=df, x='MARRIAGE', hue='DEFAULT')
marriage.set_xticklabels(['Married', 'Single', 'Other'])
plt.show()
```

```python
pd.crosstab(df.MARRIAGE, df.DEFAULT, normalize=True)
```

```python
sns.set(rc={'figure.figsize': (20, 16)})
sns.countplot(data=df, x='PAY_1', hue='DEFAULT')
plt.show()
```

```python
pd.crosstab(df.PAY_1, df.DEFAULT, margins=True)
```

```python
#Balance vs Default
sns.catplot(data=df, x='DEFAULT', y='LIMIT_BAL', jitter=True)
```

```python
#evaluating the relationship between the age and default columns
pd.crosstab(df.AGE, df.DEFAULT)
```
