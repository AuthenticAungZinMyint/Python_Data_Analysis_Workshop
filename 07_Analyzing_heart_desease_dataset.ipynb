```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("./datasets/heart.csv")
df.head()
```

```python
df.rename(  index= str, 
            columns= { 'cp': 'chest_pain',
                        'trestbps': 'rest_bp',
                        'fbs': 'fast_bld_sugar',
                        'restecg': 'rest_ecg',
                        'thalach': 'max_hr',
                        'exang': 'ex_angina',
                        'oldpeak': 'st_depr',
                        'ca': 'colored_vessels',
                        'thal': 'thalassemia'}, inplace=True)
```

```python
#plotting the distribution and relationship between specific features
sns.set(palette='pastel', rc= {'figure.figsize': (16, 10), "axes.titlesize": 18, "axes.labelsize": 16, "xtick.labelsize": 14, "ytick.labelsize": 16})
```

```python
g = sns.countplot(data=df, x='age')
g.set_title("Distribution of Age")
plt.xlabel('Age')
plt.show()
```

```python
#target column
df.target.value_counts()
```

```python
a = sns.countplot(data=df, x='target')
a.set_title('Distribution of presence of heart disease')
a.set_xticklabels(['absent', 'present'])
plt.xlabel('Present of Heart Disease')
plt.show()
```

```python
#sex columns
df.sex.value_counts()
```

```python
b = sns.countplot(data=df, x='target', hue='sex')
b.set_title("Distribution of presence of heart disease by sex")
b.legend(['female', 'male'])
b.set_xticklabels(['absent', 'persent'])
plt.show()
```

```python
#Distributions and relationship between columns with respect to the target columns
#chest pain
df.chest_pain.value_counts()
```

```python
cp = sns.countplot(data=df, hue='target', x='chest_pain')
cp.set_title("Distribution of Presence of heart disease by chest pain type")
cp.legend(['Absent', 'Present'])
cp.set_xticklabels(['Typical Anginal', 'Atypical Anginal', 'Non-aginal Pain', 'Asymptomatic'])
plt.show()
```

```python
df.colored_vessels.value_counts()
```

```python
cv = sns.countplot(data=df, hue='target', x='colored_vessels')
cv.set_title("Distribution of Presence of heart disease by Number of major vessels colored by Fluoroscopy")
cv.legend(['Absent', 'Present'])
plt.show()
```

```python
#slope
df.slope.value_counts()
```

```python
s = sns.countplot(data=df, hue='target', x='slope')
s.set_title("Distribution of Presence of heart disease by chest pain type")
s.legend(['Absent', 'Present'])
s.set_xticklabels(['Upsloping', 'Flat', 'Downsloping'])
plt.show()
```

```python
#plotting the relationship between the prescence of heart disease and maximum recorded heart rate
sns.set(palette='colorblind', style='whitegrid', rc={"figure.figsize": (12,8), "axes.titlesize": 18, "axes.labelsize": 16, "xtick.labelsize": 16, "ytick.labelsize": 16})
```

```python
f = sns.scatterplot(data=df, x='age', y='max_hr', hue='target', style='target')
f.set_title("Presence of heart disease based on age and maximum heart rate")
plt.xlabel('Age')
plt.ylabel('Mazimum heart rate')
plt.show()
```

```python
df['age_category'] = pd.cut(df.age, bins = list(np.arange(25, 85, 5)))
```

```python
plt.subplot(121)
```

```python
plt.subplot(121)
df[df.target == 1].groupby('age_category')['age'].count().plot(kind='bar')
plt.title('Present')
plt.xlabel('Age group')
plt.ylabel('Count')
```

```python
plt.subplot(122)
df[df.target == 0].groupby('age_category')['age'].count().plot(kind='bar')
plt.title('Absent')
plt.xlabel('Age group')
plt.ylabel('count')
```

```python
#Plotting the Relationship between the Presence of Heart Disease and the Cholesterol Column 

c = sns.scatterplot(data=df, x='age', y='chol', hue='target', style='target')
c.set_title('Present of heart disease based on age and cholesterol')
c.set_xlabel('Age')
c.set_ylabel('Cholestrol')
plt.show()
```

```python
df['chol_cat'] = pd.cut(df.chol, bins=list(np.arange(120, 380, 20)))
```

```python
plt.subplot(121)
df[df.target == 1].groupby('chol_cat')['chol'].count().plot(kind='bar')
plt.title('Present')
plt.xlabel("Cholestrol group")
plt.ylabel('count')
```

```python
plt.subplot(122)
df[df.target == 0].groupby('chol_cat')['chol'].count().plot(kind='bar')
plt.title('Absent')
plt.xlabel("Cholestrol group")
plt.ylabel('count')
```

```python
#observing correlations with heatmap
f, ax = plt.subplots(figsize=(12,12))
sns.heatmap(df.corr(), annot = True, linewidths= .5, fmt= '.1f', ax=ax)
plt.show()
```
